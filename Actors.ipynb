```
#O método Table scatter desenha um gráfico de dispersão que consiste em um ponto para cada linha da tabela. O primeiro argumento é o rótulo da coluna a ser plotada no eixo horizontal e o segundo argumento é o rótulo da coluna na vertical.

actors.scatter('Number of Movies', 'Total Gross')
```
